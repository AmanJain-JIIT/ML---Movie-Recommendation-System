```python
#importing libraries
from ast import Global
from textwrap import wrap
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from tkinter import *
from tkinter import ttk


#reading dataset
movies_df = pd.read_csv('movies.csv')

#data pre-processing
selected_features = ['genres', 'keywords', 'overview', 'tagline', 'title', 'cast', 'director']
for feature in selected_features:
    movies_df[feature] = movies_df[feature].fillna('')
combined_features = movies_df['genres'] + ' ' + movies_df['keywords'] + ' ' + movies_df['overview'] + ' ' + movies_df['tagline'] + ' ' + movies_df['title'] + ' ' + movies_df['cast'] + ' ' + movies_df['director']
#print(combined_features)

#converting textual data to numeric features
Vectorizer = TfidfVectorizer()
features = Vectorizer.fit_transform(combined_features)
#print(features)

#cosine similarity
similarity = cosine_similarity(features)
#print(similarity)

#recommender function
def recommender(movie_name):
    recommended = []
    list_of_movies = movies_df['title'].tolist()
    close_matches = difflib.get_close_matches(movie_name, list_of_movies)
    movie = close_matches[0]
    index = movies_df[movies_df.title == movie]['index'].values[0]
    similarity_score = list(enumerate(similarity[index]))
    sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
    i=1
    for movie in sorted_similar_movies:
        #movie_title = movies_df[movies_df.index == movie[0]]['title'].values[0]
        if (i<21):
            recommended.append(movie[0])
            i+=1

    
    return recommended
       
# defining main window
root = Tk()
root.title("Movie Recommender")
root.state("zoomed")
root.configure(bg = "black")

#find movies function
recommendations = []
def find():
    global choice
    global recommendations
    global L
    global Overview
    global L2
    
    c = choice.get()
    choice.delete(0, END)
    recommendations = recommender(c)
    print(recommendations)

    #defining new window
    new_window = Toplevel(root)
    new_window.title("Recommendations")
    new_window.state("zoomed")

    new_window.configure(bg = "black")

    frame1 = LabelFrame(new_window, padx = 20, pady = 20, bg = "red")
    frame1.pack(padx = 20, pady = 20)

    #defining prev next functions
    def prev_movie(movienumber):
        global L
        global Overview
        global L2

        L.grid_forget()
        L2.grid_forget()

        status_bar = Label(frame1, text = "Movie " + str(movienumber) + " of " + '20', bd = 2, relief = SUNKEN, padx = 5, pady = 5, bg = "black", fg = "white")
        status_bar.configure(font=("bold", 20))
        status_bar.grid(row = 0, column = 1, sticky = W + E, padx = 10, pady = 10)  

        prev_ = Button(frame1, text = "<<",bg = "white", fg = "black", borderwidth = 5, padx = 20, pady = 20, command=lambda : prev_movie(movienumber-1))
        prev_.configure(font = (20))

        next_ = Button(frame1, text = ">>",bg = "white", fg = "black", borderwidth = 5, padx = 20, pady = 20, command = lambda : next_movie(movienumber+1))
        next_.configure(font = (20))

        prev_.grid(row = 2, column = 0, padx = 20, pady = 20)
        next_.grid(row = 2, column = 2, padx = 20, pady = 20) 

        i = movienumber-1
        title =  movies_df[movies_df.index == recommendations[i]]['title'].values[0]
        director = movies_df[movies_df.index == recommendations[i]]['director'].values[0]
        genres = movies_df[movies_df.index == recommendations[i]]['genres'].values[0]
        cast = movies_df[movies_df.index == recommendations[i]]['cast'].values[0]
        overview = movies_df[movies_df.index == recommendations[i]]['overview'].values[0]

        movie_data = "Title : " + title + '\n' + "Director : " + director + "\n" + "Genres : " + genres

        
        L = Label(frame1, text = movie_data, padx = 20, pady = 20, bg = "red", fg = "white", font = ("Times", 30, "bold"))
        L.grid(row = 1, column = 1)

        Overview = Label(frame1, text = "OVERVIEW : ", padx = 20, pady = 20, bg = "red", fg = "black", font = ("Times", 40, "bold"))
        Overview.grid(row = 2, column = 1, pady = (20, 0))

        L2 = Label(frame1, text = overview, wraplength= 500 ,padx = 20, pady = 20, bg = "red", fg = "white", font = ("Times", 20, "bold"))
        L2.grid(row = 3, column = 1, pady = 0)
        

        if(movienumber == 1):
            prev_ = Button(frame1, text = "<<",bg = "white", fg = "black", borderwidth = 5, padx = 20, pady = 20, command=lambda : prev_movie(movienumber-1), state=DISABLED)
            prev_.configure(font = (20))
            prev_.grid(row = 1, column = 0, padx = 20, pady = 20)

    
    def next_movie(movienumber):
        global L
        global Overview
        global L2

        L.grid_forget()
        L2.grid_forget()

        status_bar = Label(frame1, text = "Movie " + str(movienumber) + " of " + '20', bd = 2, relief = SUNKEN, padx = 5, pady = 5, bg = "black", fg = "white")
        status_bar.configure(font=("bold", 20))
        status_bar.grid(row = 0, column = 1, sticky = W + E, padx = 10, pady = 10)  

        prev_ = Button(frame1, text = "<<",bg = "white", fg = "black", borderwidth = 5, padx = 20, pady = 20, command=lambda : prev_movie(movienumber-1))
        prev_.configure(font = (20))

        next_ = Button(frame1, text = ">>",bg = "white", fg = "black", borderwidth = 5, padx = 20, pady = 20, command = lambda : next_movie(movienumber+1))
        next_.configure(font = (20))

        prev_.grid(row = 2, column = 0, padx = 20, pady = 20)
        next_.grid(row = 2, column = 2, padx = 20, pady = 20) 

        i = movienumber-1
        title =  movies_df[movies_df.index == recommendations[i]]['title'].values[0]
        director = movies_df[movies_df.index == recommendations[i]]['director'].values[0]
        genres = movies_df[movies_df.index == recommendations[i]]['genres'].values[0]
        cast = movies_df[movies_df.index == recommendations[i]]['cast'].values[0]
        overview = movies_df[movies_df.index == recommendations[i]]['overview'].values[0]

        movie_data = "Title : " + title + '\n' + "Director : " + director + "\n" + "Genres : " + genres
        
        L = Label(frame1, text = movie_data, padx = 20, pady = 20, bg = "red", fg = "white", font = ("Times", 30, "bold"))
        L.grid(row = 1, column = 1)

        Overview = Label(frame1, text = "OVERVIEW : ", padx = 20, pady = 20, bg = "red", fg = "black", font = ("Times", 40, "bold"))
        Overview.grid(row = 2, column = 1, pady = (20, 0))

        L2 = Label(frame1, text = overview, wraplength= 500 ,padx = 20, pady = 20, bg = "red", fg = "white", font = ("Times", 20, "bold"))
        L2.grid(row = 3, column = 1, pady = 0)

        if(movienumber==20):
            next_ = Button(frame1, text = ">>",bg = "white", fg = "black", borderwidth = 5, padx = 20, pady = 20, command = lambda : next_movie(movienumber+1), state=DISABLED)
            next_.configure(font = (20))
            next_.grid(row = 1, column = 2, padx = 20, pady = 20) 


        


    # defining display of second window
    status_bar = Label(frame1, text = "Movie 1 of " + '20', bd = 2, relief = SUNKEN, padx = 5, pady = 5, bg = "black", fg = "white")
    status_bar.configure(font=("bold", 20))
    status_bar.grid(row = 0, column = 1, sticky = W + E, padx = 10, pady = 10)  
    
    prev_ = Button(frame1, text = "<<",bg = "white", fg = "black", borderwidth = 5, padx = 20, pady = 20, command=lambda : prev_movie(), state = DISABLED)
    prev_.configure(font = (20))
    
    next_ = Button(frame1, text = ">>",bg = "white", fg = "black", borderwidth = 5, padx = 20, pady = 20, command = lambda : next_movie(2))
    next_.configure(font = (20))
    
    prev_.grid(row = 2, column = 0, padx = 20, pady = 20)
    next_.grid(row = 2, column = 2, padx = 20, pady = 20)  

    i = 0
    title =  movies_df[movies_df.index == recommendations[i]]['title'].values[0]
    director = movies_df[movies_df.index == recommendations[i]]['director'].values[0]
    genres = movies_df[movies_df.index == recommendations[i]]['genres'].values[0]
    cast = movies_df[movies_df.index == recommendations[i]]['cast'].values[0]
    overview = movies_df[movies_df.index == recommendations[i]]['overview'].values[0]

    movie_data = "Title : " + title + '\n' + "Director : " + director + "\n" + "Genres : " + genres
    
    L = Label(frame1, text = movie_data, padx = 20, pady = 20, bg = "red", fg = "white", font = ("Times", 30, "bold"))
    L.grid(row = 1, column = 1)

    Overview = Label(frame1, text = "OVERVIEW : ", padx = 20, pady = 20, bg = "red", fg = "black", font = ("Times", 40, "bold"))
    Overview.grid(row = 2, column = 1, pady = (20, 0))

    L2 = Label(frame1, text = overview, wraplength= 500 ,padx = 20, pady = 20, bg = "red", fg = "white", font = ("Times", 20, "bold"))
    L2.grid(row = 3, column = 1, pady = 0)

#button functions
def on_enter(e):
   button.config(background='black', foreground= "white")

def on_leave(e):
   button.config(background= '#454545', foreground= 'white')

#defining frame
frame = LabelFrame(root, padx = 20, pady = 20, bg = "red")
frame.pack(padx = 20, pady = 20)

#defining app header
heading = Label(frame, text = "MOVIE RECOMMENDER", bg = "red")
heading.configure(font = ('Times',80, "bold"))
heading.grid(row = 0, column = 0, columnspan = 3, padx = 40, pady = 40)

Label(frame, text = "ENTER YOUR FAVOURITE MOVIE NAME", padx = 20, pady = 20, bg = "red", fg = "white", font = ("Times", 30, "bold")).grid(row = 1, column = 1, padx = 20, pady = 30)

choice = Entry(frame, width = 50, borderwidth = 5)
choice.configure(font = ("Times", 30, "bold"))
choice.grid(row = 2, column = 1, padx = 10, pady = 10, ipadx = 30, ipady = 10)

button = Button(frame, text = "FIND MORE MOVIES", padx = 10, pady = 10, borderwidth = 10, relief = RAISED,bg="#454545", fg="white", command = find)
button.configure(font = ("Times", 20, "bold"))
button.grid(row = 3, column = 1, padx = 10, pady = (50, 10), ipadx = 40, ipady = 5)

button.bind('<Enter>', on_enter)
button.bind('<Leave>', on_leave)

# mainloop which run infinitely
root.mainloop()
```
